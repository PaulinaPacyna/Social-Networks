## Own functions

```python
import matplotlib.pyplot as plt
import numpy as np
```

1. Draw square lattice with your own function. 

```python
N=5
M=4
# lines will be denoted as ((x_start, x_end), (y_start, y_end))
vertical_edges = [((i, i), (0, M)) for i in range(N+1)]  # we manipulate the x values
horizontal_edges = [((0, N), (i,i)) for i in range(M+1)] # we manipulate the y values
vertices = [(i, j) for i in range(N+1) for j in range(M+1)] # cartesian product of {0, .., N} x {0, .., M} 

fig, ax = plt.subplots()
for e in horizontal_edges + vertical_edges: # draw all edges
    ax.plot(*e, c='b')
for v in vertices:
    ax.plot(*v, 'p', c='r')
```

2. Draw chain with your own function. 

```python
K=8
edges = [range(K), range(K)] # x coordinates in the first argument, y coordinates in the second argument
plt.plot(*edges) #lines
plt.plot(*edges, 'p', c='r') #points
```

3. Draw complete graph with your own function. 

```python
K=6
vertices = [(np.random.uniform(), np.random.uniform()) for _ in range(K)] # choosing randomly K points
edges = [zip(i, j) for i in vertices for j in vertices if i!=j] # connecting all vertices
for e in edges:
    plt.plot(*e, c='b')
for v in vertices:
    plt.plot(*v, 'p', c='r')
```
